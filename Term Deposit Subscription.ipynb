### Problem Statement

The dataset pertains to direct marketing campaigns conducted by a Portuguese banking institution, where the primary mode of communication was through phone calls. The objective of these campaigns was to determine whether clients would subscribe to a bank term deposit, with the **outcome variable denoted as 'y'** and having two possible values: **'yes'** indicating **subscription** and **'no'** indicating **non-subscription.**

### Dataset Description

1. **Age:** 
    - Age of the client. This feature represents the client's age, providing insight into the demographic profile of individuals targeted in the marketing campaigns.

2. **job:** 
    - Occupation of the client. Describes the client's profession or employment, offering information on their economic status and potential financial needs.

3. **marital:** 
    - Marital status of the client. Indicates whether the client is single, married, divorced, or in another marital status, providing context on their personal life.

4. **education:** 
    - Level of education of the client. Reflects the educational background of the client, influencing their financial decision-making and product preferences.

5. **default:** 
    - Whether the client has credit in default (yes/no). Indicates if the client has defaulted on credit, providing insights into their financial stability and creditworthiness.

6. **housing:** 
    - Whether the client has a housing loan (yes/no). Informs whether the client has an existing housing loan, impacting their financial commitments and capacity for additional investments.

7. **loan:** 
    - Whether the client has a personal loan (yes/no). Indicates if the client has a personal loan, affecting their financial obligations and capacity for new financial commitments.

8. **contact:**
    - Type of communication contact. Describes the method of communication, such as cellular or telephone, used in the marketing campaigns to reach the client.

9. **month:** 
    - Last contact month of the year. Specifies the month when the last contact with the client occurred, offering temporal information for campaign analysis.

10. **day_of_week:** 
    - Last contact day of the week. Indicates the day of the week when the last contact with the client took place, aiding in identifying potential patterns in response based on weekdays.

11. **duration:** 
    - Duration of the last contact in seconds. Influences the outcome target; a duration of 0 typically results in 'no' subscription, providing a crucial temporal aspect to the analysis.

12. **campaign:** 
    - Number of contacts performed during this campaign for this client. Represents the frequency of contacts made with the client during the current marketing campaign.

13. **pdays:** 
    - Number of days that passed by after the client was last contacted from a previous campaign. A value of 999 signifies no previous contact, while other values indicate the duration since the last interaction.

14. **previous:** 
    - Number of contacts performed before this campaign for this client. Indicates the historical engagement level with the client in previous marketing campaigns.

15. **emp.var.rate:** 
    - Employment variation rate. Reflects changes in employment conditions, offering an economic indicator that may influence client decisions.

16. **cons.price.idx:** 
    - Consumer price index. Provides information on the overall price levels of goods and services, influencing consumer purchasing power.

17. **cons.conf.idx:** 
    - Consumer confidence index. Indicates the level of confidence consumers have in the economy, affecting their willingness to make financial commitments.

18. **euribor3m:**
    - Euribor 3-month rate. Represents the interest rate at which European banks offer three-month loans to each other, influencing financial market conditions.

19. **nr.employed:** 
    - Number of employees. Reflects the total number of employees, serving as an economic indicator with potential impacts on consumer behavior.

20. **poutcome:** 
    - Outcome of the previous marketing campaign. Describes the result of the previous campaign, providing historical context for understanding client responses.

21. **y(Target Variable):** 
    - The target variable indicating whether the client subscribed to the term deposit (yes/no). This binary outcome determines the success of the marketing campaign in terms of term deposit subscriptions.

## Model Building

```python
rf =  RandomForestClassifier()
dtr = DecisionTreeClassifier()
knn = KNeighborsClassifier()
lr = LogisticRegression()
svc = SVC()

models = [rf, dtr, knn, lr, svc]
metrics_list = []

for model in models:
    model.fit(x_train, y_train)
    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)

    accuracy_train = metrics.accuracy_score(y_train, y_pred_train)
    accuracy_test = metrics.accuracy_score(y_test, y_pred_test)
    accuracy_metrics = {'Algorithm'    : str(model).split('(')[0],
                        'Accuracy_Train': accuracy_train,
                        'Accuracy_Test' : accuracy_test}

    metrics_list.append(accuracy_metrics)
```

```python
Best_cls_model =pd.DataFrame(metrics_list)
Best_cls_model
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_features': [None, 'sqrt', 'log2'],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 4, 8],
}

rf = RandomForestClassifier()
x_train, x_test, y_train, y_test = train_test_split( x_train_resample, y_train_resample, test_size=0.30, random_state=42)

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5,n_jobs=-1, verbose=3)
grid_search.fit(x_train, y_train)

print("Best Cross-validated Parameters: ", grid_search.best_params_)
print("Best Cross-validated Score: ", grid_search.best_score_)
```

**Interpretation of Confusion Matrix:**

**Accuracy:** The model has an overall accuracy of 93%, indicating that it correctly predicts the class for 93% of the instances in the dataset.

**Precision:** The precision for class 0 is 96%, and for class 1, it's 88%. This means that, among the instances predicted as positive, 96% (for class 0) and 88% (for class 1) were actually positive.

**Recall (Sensitivity):** The recall for class 0 is 94%, and for class 1, it's 91%. This means that the model correctly identified 94% of the actual positive instances for class 0 and 91% for class 1.

**F1-Score:** The F1-score is the harmonic mean of precision and recall. It balances precision and recall. The weighted average of the F1-scores is 0.93.

**Support:** The support column in the classification report shows the number of actual instances for each class.
